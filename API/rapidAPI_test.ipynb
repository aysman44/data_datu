```python
import os
import http.client
import json
import pandas as pd
from dotenv import load_dotenv
load_dotenv()

def rapidAPI_call(url):
    """
    Returns a dictionary of response data details depending on the rapidAPI url
    url = GET api enpoint
    Host = specific request rapidAPI url
    X_Key = assigned rapidAPI key for specific endpoing i.e. flights, flightDestination, hotels etc.
    X_host = usually the same as Host
    """
    secrets = json.loads(os.getenv("rapid_api_secrets"))
    X_key = secrets['X-Rapidapi-Key']
    X_host = secrets['X-Rapidapi-Host']
    Host = secrets['Host']
    conn = http.client.HTTPSConnection(f"{Host}")

    # Headers including the API key
    headers = {
        "X-Rapidapi-Key": f"{X_key}",
        "X-Rapidapi-Host": f"{X_host}"
    }

    conn.request("GET", f"{url}", headers=headers)
    response = conn.getresponse()
    data = response.read()
    dict = json.loads(data)
    #df = pd.DataFrame((dict['data']))
    return dict

#print(data.decode("utf-8"))
```

```python
secrets = json.loads(os.getenv("rapid_api_secrets"))
dest_url = secrets['flight_dest_url']
```

```python
dict_dest = rapidAPI_call(dest_url)
dict_dest
```
